```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter

# Define the base directory for results
Results_dir = os.path.join('..', 'Results')
CSV_Results_dir = os.path.join(Results_dir, "LLM_AFR", "Journal_paper", "CSV_results" )
csv_path = os.path.join(CSV_Results_dir, 'prediction_vs_ground_truth_results_all.csv')

# Load the CSV file
data = pd.read_csv(csv_path)
```

### No. of features in CAD Files

```python
# Remove the content inside parentheses from the feature names
data['feature_name'] = data['feature_name'].str.replace(r'\s*\(.*\)', '', regex=True)

# Group the data by the cleaned feature_name and sum the gt_quantity for each feature
feature_counts = data.groupby('feature_name')['gt_quantity'].sum().reset_index()

# Sorting the feature counts from highest to lowest
feature_counts_sorted = feature_counts.sort_values(by='gt_quantity', ascending=False)

# Function to format the Y-axis
def format_y_axis(x, pos):
    return f'{x:.0f}'

# Plotting the bar chart with compact spacing between bars and feature names
figure_width = 10
figure_height = 8
palette = sns.color_palette("Set2", len(feature_counts_sorted))

plt.figure(figsize=(figure_width, figure_height))
bars = sns.barplot(data=feature_counts_sorted, x='feature_name', y='gt_quantity', palette=palette, dodge=False)
plt.xlabel('Feature Name', fontsize=20, color='black', labelpad=10)
plt.ylabel('Number of Features', fontsize=20, color='black', labelpad=10)
plt.title('No. of Each Feature in All CAD Files', fontsize=22, color='black', pad=7)
plt.xticks(rotation=45, fontsize=16, color='black', ha='right', rotation_mode='anchor')
plt.yticks(fontsize=16, color='black')
plt.grid(True, linestyle=':', linewidth=1, alpha=0.8)

# Add contour to the bars
for bar in bars.patches:
    bar.set_edgecolor('black')
    bar.set_linewidth(1.2)

current_fig_width = plt.gcf().get_size_inches()[0]
bar_count = len(feature_counts_sorted)
spacing_reduction = 0.05  # Reduce this value to decrease spacing between bars
new_fig_width = current_fig_width - spacing_reduction * (bar_count - 1)

plt.gcf().set_size_inches(new_fig_width, figure_height)

plt.tight_layout(pad=2.0)

plt.gca().yaxis.set_major_formatter(FuncFormatter(format_y_axis))

plt.savefig(
            os.path.join(Results_dir, "LLM_AFR", "Visualization", "feature_counts_bar_chart", "feature_counts_bar_chart_overall.png"),
            dpi=600,  
            bbox_inches='tight'  
        )

plt.show()
```

### Combined Plot for Comparison (Revised Manuscript)

```python
import pandas as pd
import os
import matplotlib.pyplot as plt

# Define the base directory for results
Results_dir = os.path.join('..', 'Results')
CSV_Results_dir = os.path.join(Results_dir, "LLM_AFR", "Journal_paper", "CSV_results")
csv_path = os.path.join(CSV_Results_dir, 'prediction_vs_ground_truth_results_all.csv')

# Load the CSV file
data = pd.read_csv(csv_path)

# Remove the content inside parentheses from the feature names
data['feature_name'] = data['feature_name'].str.replace(r'\s*\(.*\)', '', regex=True)

# Classify parts into easy, medium, and hard based on their names
data['complexity'] = data['part_name'].apply(lambda x: 'easy' if x.startswith('easy') else 
                                                           ('medium' if x.startswith('medium') else 'hard'))

# Group data by feature_name and complexity and sum the quantities
grouped_data = data.groupby(['feature_name', 'complexity'])['gt_quantity'].sum().reset_index()

# Pivot data for easier plotting, reorder columns to have 'easy', 'medium', 'hard'
pivot_data = grouped_data.pivot(index='feature_name', columns='complexity', values='gt_quantity').fillna(0)
pivot_data = pivot_data[['easy', 'medium', 'hard']]  # Reordering the columns

# Sorting features by the sum across all complexity levels
pivot_data['total'] = pivot_data.sum(axis=1)
pivot_data = pivot_data.sort_values(by='total', ascending=False).drop(columns='total')

# Plotting grouped bar chart with improved visual settings
figure_width = 14
figure_height = 8
ax = pivot_data.plot(kind='bar', figsize=(figure_width, figure_height), width=0.85, edgecolor='black')

# Improve visualization with better contrast and thicker bars
for bar in ax.containers:
    for patch in bar:
        patch.set_linewidth(1.5)

plt.xlabel('Feature Name', fontsize=20, color='black', labelpad=10)
plt.ylabel('Number of Features', fontsize=20, color='black', labelpad=10)
plt.title('Comparison of Feature Counts Across Complexity Levels', fontsize=22, color='black', pad=7)

plt.xticks(rotation=45, fontsize=16, color='black', ha='right')
plt.yticks(fontsize=16, color='black')
plt.grid(True, linestyle=':', linewidth=1, alpha=0.8)

# Adjust legend order and font sizes
handles, labels = ax.get_legend_handles_labels()
ordered_handles = [handles[labels.index('easy')], handles[labels.index('medium')], handles[labels.index('hard')]]
ordered_labels = ['Easy', 'Medium', 'Hard']

legend = plt.legend(ordered_handles, ordered_labels, title="Complexity Level", fontsize=12, title_fontsize=14, loc='upper right', bbox_to_anchor=(0.98, 0.98))

# Set y-axis tick spacing with an interval of 25
plt.yticks(range(0, int(pivot_data.max().max()) + 25, 25))

# Save the high-resolution plot
plt.savefig(
    os.path.join(Results_dir, "LLM_AFR", "Journal_paper", "Visualization","feature_counts_bar_chart", "feature_counts_comparison.png"),
    dpi=300,  
    bbox_inches='tight'  
)

plt.show()
```

```python
import pandas as pd
import os
import matplotlib.pyplot as plt

# Define the base directory for results
Results_dir = os.path.join('..', 'Results')
CSV_Results_dir = os.path.join(Results_dir, "LLM_AFR", "Journal_paper", "CSV_results")
csv_path = os.path.join(CSV_Results_dir, 'prediction_vs_ground_truth_results_all.csv')

# Load the CSV file
data = pd.read_csv(csv_path)

# Remove the content inside parentheses from the feature names
data['feature_name'] = data['feature_name'].str.replace(r'\s*\(.*\)', '', regex=True)

# Classify parts into easy, medium, and hard based on their names
data['complexity'] = data['part_name'].apply(lambda x: 'easy' if x.startswith('easy') else 
                                                           ('medium' if x.startswith('medium') else 'hard'))

# Group data by feature_name and complexity and sum the quantities
grouped_data = data.groupby(['feature_name', 'complexity'])['gt_quantity'].sum().reset_index()

# Pivot data for easier plotting, reorder columns to have 'easy', 'medium', 'hard'
pivot_data = grouped_data.pivot(index='feature_name', columns='complexity', values='gt_quantity').fillna(0)
pivot_data = pivot_data[['easy', 'medium', 'hard']]  # Reordering the columns

# Exclude the feature 'Rib' from the dataset
pivot_data_filtered = pivot_data.drop(index='Rib', errors='ignore')

# Sorting features by the sum across all complexity levels
pivot_data_filtered['total'] = pivot_data_filtered.sum(axis=1)
pivot_data_filtered = pivot_data_filtered.sort_values(by='total', ascending=False).drop(columns='total')

# Plotting grouped bar chart with improved visual settings after excluding 'Rib'
figure_width = 14
figure_height = 8
ax = pivot_data_filtered.plot(kind='bar', figsize=(figure_width, figure_height), width=0.85, edgecolor='black')

# Improve visualization with better contrast and thicker bars
for bar in ax.containers:
    for patch in bar:
        patch.set_linewidth(1.5)

# Add data labels to ensure visibility of small values
for bar in ax.containers:
    ax.bar_label(bar, fmt='%d', fontsize=10, padding=3, color='black', rotation=90, label_type='edge')

plt.xlabel('Feature Name', fontsize=20, color='black', labelpad=10)
plt.ylabel('Number of Features', fontsize=20, color='black', labelpad=10)
plt.title('Comparison of Feature Counts Across Complexity Levels (Excluding Rib)', fontsize=22, color='black', pad=7)

plt.xticks(rotation=45, fontsize=16, color='black', ha='right')
plt.yticks(fontsize=16, color='black')
plt.grid(True, linestyle=':', linewidth=1, alpha=0.8)

# Adjust legend order and font sizes
handles, labels = ax.get_legend_handles_labels()
ordered_handles = [handles[labels.index('easy')], handles[labels.index('medium')], handles[labels.index('hard')]]
ordered_labels = ['Easy', 'Medium', 'Hard']

legend = plt.legend(ordered_handles, ordered_labels, title="Complexity Level", fontsize=12, title_fontsize=14, loc='upper right', bbox_to_anchor=(0.98, 0.98))

# Set y-axis tick spacing with an interval of 25
plt.yticks(range(0, int(pivot_data_filtered.max().max()) + 25, 25))

# Save the high-resolution plot excluding Rib
plt.savefig(
    os.path.join(Results_dir, "LLM_AFR", "Journal_paper", "Visualization", "feature_counts_bar_chart", "feature_counts_excluding_rib.png"),
    dpi=300,  
    bbox_inches='tight'  
)

plt.show()
```
